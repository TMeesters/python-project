## Data Cleanup

Now that we loaded the data from switchup website, we need to clean data and prepare them to insert into
database. 

Activities:
* Remove HTML like codes from fields such as Description
* Check for NULL values and decide what to do with them

```python
#imports
import pandas as pd
import re
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
!pip install geopy
import folium
from geopy.geocoders import Nominatim
from folium.plugins import MarkerCluster

from sklearn.impute import SimpleImputer
```

```python
# Load CSV files we downloaded, in order to clean them
badges = pd.read_csv('badges.csv')
comments = pd.read_csv('comments.csv')
schools = pd.read_csv('schools.csv')
locations = pd.read_csv('locations.csv')
courses = pd.read_csv('courses.csv')
```

#### Remove HTML

```python
def strip_html(str):
    return re.sub('<[^<]+?>', '', str)
```

```python
schools['description'] = schools.apply(lambda r: strip_html(r['description']), axis=1)
schools.head()
```

```python
badges['description'] = badges.apply(lambda r: strip_html(r['description']), axis=1)
badges.head()
```

```python
comments.head()
```

```python
comments['body'] = comments.apply(lambda r: strip_html(r['body']), axis=1)
comments.head()
```

```python
comments = comments[['graduatingYear', 'jobTitle', 'tagline', 'program', 'overallScore', 'overall', 'curriculum', 'jobSupport', 'review_body', 'school']]
comments.head()
```

```python
#  Comments - Mean of OverallScore of each school 

# Sort the DataFrame by 'school' and 'overallScore' in descending order
comments_sorted = comments.sort_values(by=['school', 'overallScore'], ascending=[True, False])

# Reset index to move 'school' from index to a regular column
comments_sorted = comments_sorted.reset_index()

# Group by 'school' and calculate the mean overall score for each school
school_overall_scores = comments_sorted.groupby('school')['overallScore'].mean().reset_index()

# Sort the result by mean overall score in descending order
school_overall_scores = school_overall_scores.sort_values(by='overallScore', ascending=False)

# Display the result
print(school_overall_scores)

```

```python
plt.figure(figsize=(12, 6))
plt.bar(school_overall_scores['school'], school_overall_scores['overallScore'], color='skyblue')
plt.xlabel('School')
plt.ylabel('Mean Overall Score')
plt.title('Mean Overall Score of Each School')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a new DataFrame with 'country.name' and 'school'
new_dfc = locations[['country.name', 'school']].copy()
# Replace NaN values in 'country.name' with 'Online'
new_dfc['country.name'].fillna('Online', inplace=True)

# Display the new DataFrame
print(new_dfc)

# Function to get latitude and longitude for a given location
def get_lat_long(location):
    geolocator = Nominatim(user_agent="my_geocoder")
    location_info = geolocator.geocode(location)
    return location_info.latitude, location_info.longitude

# Apply the function to get latitude and longitude for each row in the DataFrame
new_dfc['coordinates'] = new_dfc['country.name'].apply(lambda x: get_lat_long(x) if x != 'Online' else (0, 0))

# Create a folium map centered on the first location
m = folium.Map(location=[new_dfc['coordinates'].iloc[0][0], new_dfc['coordinates'].iloc[0][1]], zoom_start=4)

# Create a MarkerCluster for better visualization
marker_cluster = MarkerCluster().add_to(m)

# Define a color dictionary for each school
color_dict = {
    'udacity': 'blue', 'the-tech-academy': 'green', 'clarusway': 'red',
    'brainstation': 'purple', 'springboard': 'orange', 'knowledgehut': 'pink',
    'coding-dojo': 'brown', 'ironhack': 'gray', 'app-academy': 'cyan',
    'tripleten': 'lightblue', 'codesmith': 'darkgreen', 'coding-temple': 'darkred',
    'nucamp': 'darkpurple', 'learningfuze': 'darkorange', '4geeks-academy': 'darkpink',
    'altcademy': 'black', 'actualize': 'lightgreen', 'skillcrush': 'lightgray',
    'codeworks': 'lightcyan', 'code-platoon': 'lightyellow', 'devcodecamp': 'lightcoral',
    'tech-elevator': 'lightpink', 'tech-i-s': 'lightviolet', 'hack-reactor': 'lightbrown'
}

# Add markers for each location with different colors
for index, row in new_dfc.iterrows():
    school_color = color_dict.get(row['school'], 'gray')  
    folium.Marker(location=row['coordinates'], popup=row['school'], icon=folium.Icon(color=school_color)).add_to(marker_cluster)

# Add custom HTML legend
legend_html = """
     <div style="position: fixed;
                 top: 10px; left: 10px; width: 200px; height: 550px;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color:white; padding: 10px;">
     &nbsp; <strong>Legend</strong> <br>
     """
for school, color in color_dict.items():
    legend_html += f'&nbsp; <i class="fa fa-circle" style="color:{color}"></i>&nbsp; {school} <br>'
legend_html += """
     </div>
     """
m.get_root().html.add_child(folium.Element(legend_html))

# Display the map
m

# # Save the map 
# m.save("map.html")
```
