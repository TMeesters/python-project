```python


# define rooms and items
# ALVARO: adding new items missing in the initial sample code. 
# TARINEE: Delete extra spaces in dictionaries to make it better readable. 
# TARINEE: re-arrange items per rooms 
# TARINEE: addding color code for strip (answering box) for intended_action

BLUE = '\033[94m'
ENDC = '\033[0m'


# TARINEE: I put definition of all doors before defining each items per rooms because door d is being mentioned in the code in bedroom 2
# when we run codes in bedroom 2, we will get an error becuase door d is not being defined yet.  
door_a = {"name": "door a","type": "door",}
door_b = {"name": "door b","type": "door",}
door_c = {"name": "door c","type": "door",}
door_d = {"name": "door d","type": "door",}


# Room 1: Game Room 
# TARINEE: Add emoji üîë to each key a,b,c,d
game_room = {"name": "game room","type": "room",}
couch = {"name": "couch", "type": "furniture",}
# door_a = {"name": "door a","type": "door",}# Tarinee: this line can be deleted because we've alredy defined door a
key_a = {"name": "key for üîë door a", "type": "key", "target": door_a}
piano = {"name": "piano", "type": "furniture",}
    

# Room 2: Bedroom 1
bedroom_1 = {"name": "bedroom 1","type": "room",}
queen_bed = {"name": "queen bed","type": "furniture",}
# door_b = {"name": "door b","type": "door",}# Tarinee: this line can be deleted because we've alredy defined door b
# door_c = {"name": "door c","type": "door",}# Tarinee: this line can be deleted because we've alredy defined door c
key_b = {"name": "key for üîë door b", "type": "key", "target": door_b}

# Room 3: Bedroom 2
bedroom_2 = {"name": "bedroom 2","type": "room",}
double_bed = {"name": "double bed","type": "furniture",}
key_c = {"name": "key for door c","type": "key","target": door_c,}
dresser = {"name": "dresser","type": "furniture",}
key_d = {"name": "key for üîë door d", "type": "key", "target": door_d}

# Room 4: Living Room
# door_d = {"name": "door d","type": "door",}# Tarinee: this line can be deleted because we've alredy defined door d
living_room = {"name": "living room","type": "room",}
dining_table = {"name": "dining table","type": "furniture",}


# Target room: Outside
outside = {"name": "outside"}



# define which items/rooms are related

# ALVARO: adding new relations missing in the initial sample code.


all_rooms = [game_room, bedroom_1, bedroom_2, living_room, outside] # TARINEE: I think this line does not add anything in the code. It can be deleted

all_doors = [door_a, door_b, door_c, door_d] # TARINEE: This line can be deleted 


# TARINEE, delete extra spaces in dictionaries to make it better readable. 
object_relations = {"game room":[couch, piano, door_a],"piano": [key_a],"door a":[game_room, bedroom_1],
"bedroom 1": [door_a, queen_bed, door_b, door_c],"queen bed": [key_b],"door b" : [bedroom_1, bedroom_2],
"bedroom 2": [door_b, double_bed, dresser],"double bed": [key_c],"dresser": [key_d],"door c": [bedroom_1, living_room], 
"living room": [door_c, dining_table, door_d],"dining table": [],"door d": [living_room, outside], "outside": [door_d]}


# define game state. Do not directly change this dict.
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This
# way you can replay the game multiple times.

INIT_GAME_STATE = {"current_room": game_room, "keys_collected": [], "target_room": outside}
game_state = INIT_GAME_STATE.copy()  
    

def linebreak():
    print("\n\n")
     
    # TARINEE, instead of triple quote sign, put comment sign (#) in the front since this is not belong to the code. See below;
    # Print a line break
    
   
def start_game():
    # TARINEE: instead of triple quote sign, put comment sign(#) in the front since this is not belong to the code. See below;
    # Start the game  
    
    # ALVARO: Divided the paragraph in 3 different lines for user readibility while playing.
    print("You wake up on a couch and find yourself in a strange house with no windows which you have never been to before.")
    print("You don't remember why you are here and what had happened before.")
    print("You feel some unknown danger üí•üí• is approaching and you must get out of the house, NOW! üèÉ‚ÄçüèÉ‚ÄçüèÉ‚Äç")
    # TARINEE: add emojis üí•üí• and üèÉ‚ÄçüèÉ‚ÄçüèÉ‚Äç in the last sentence.
          
    # ALVARO: added a linebreak for user readibility while playing.
    linebreak()    
    play_room(game_state["current_room"])
 

def play_room(room):
    # TARINEE: instead of triple quote sign, put comment sign(#) in the front since they are not belong to the code. See below;
    # Play a room. First check if the room being played is the target room.
    #If it is, the game will end with success. Otherwise, let player either
    #explore (list all items in this room) or examine an item found here.    
    
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        
        #TARINEE,add upper & bold funtions and add emojis to make it more noticable.
        text = "Congrats! üéâ You escaped the room!üéà"  
        bold_text = "\033[1m" + text.upper()+ "\033[0m"
        print(bold_text)
        
        play_again = input(f"{BLUE}Do you want to play again? (yes/no):{ENDC}").strip().lower()
            if play_again == "yes":
                start_game()
            else:
                print("Thanks for playing! Goodbye.")
                return

# start_game()
        
    else:
        print("You are now in " + room["name"] + ".") # ALV: I added "." at the end.

        # ALVARO: I introduced the LOWER() method to make sure the code reads any capital letter introduced by the user.
       
        # TARINEE: Define code for blue color and then ENDC is used to reset it back to the default color for any text displayed after the input prompt.

        intended_action = input(f"{BLUE}What would you like to do? Type 'explore' or 'examine'?{ENDC}").lower().strip()


        if intended_action == "explore":
            explore_room(room)
            # ALVARO: Introducing a linebreak here, makes easier to read while playing.
            linebreak()
            play_room(room)

        elif intended_action == "examine":
            
            # ALVARO: introducing these two lines so the user has a reminder of what objects are available to be examined.
            items = [i["name"] for i in object_relations[room["name"]]]
            print("You can examine " + ", ".join(items) + ".")
            
            # ALVARO: introducing a LOWER() method to avoid capital case in the answer.
            #TARINEE: adjust the code to make the strip in blue and reset it back after the input promp
            examine_item(input(f"{BLUE}What would you like to examine?{ENDC}").lower().strip())
      
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
         
            play_room(room)
            linebreak()  
        # ALVARO: added a linebreak for user readibility while playing.

            
def explore_room(room):
    # TARINEE: instead of triple quote sign, put comment sign(#) in the front since they are not belong to the code. See below;
    # Explore a room. List all items belonging to this room.
          
    items = [i["name"] for i in object_relations[room["name"]]]
    print("You explore the room. This is " + room["name"] + ". You find " + ", ".join(items) + ".")
    

def get_next_room_of_door(door, current_room):
    # TARINEE: instead of triple quote sign, put comment sign(#) in the front since they are not belong to the code.See below;
    # From object_relations, find the two rooms connected to the given door.Return the room that is not the current room.
     
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room
        
      

def examine_item(item_name):
    # TARINEE: instead of triple quote sign, put comment sign(#) in the front since they are not belong to the code. See below;
    # Examine an item which can be a door or furniture. First make sure the intended item belongs to the current room.
    # Then check if the item is a door. Tell player if the key hasn't been collected yet.
    # Otherwise, ask the player if they want to go to the next room. If the item is not a door, then check if it contains keys.
    # Collect the key if found and update the game state. At the end, play either the current or the next room
    # depending on the game state to keep playing.        
        
    current_room = game_state["current_room"]
    next_room = ""
    output = None

    for item in object_relations[current_room["name"]]:
        if item["name"] == item_name:
            output = "You examine " + item_name + ". "
            if item["type"] == "door":
                have_key = False
                for key in game_state["keys_collected"]:
                    if key["target"] == item:
                        have_key = True
                if have_key:
                    output += "You unlock it with a key you have. ü§©"
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key."
            else:
                if item["name"] in object_relations and len(object_relations[item["name"]]) > 0:
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            # ALVARO: introducing a linebreak() for user readability while playing.
            break

    if output is None:
        print("The item you requested is not found in the current room.")
        # ALVARO: introducing a linebreak() for user readability while playing.
        
        # TARINEE: adjust the code to make the strip in blue and reset it back after the input promp     
    if next_room and input(f"{BLUE}Do you want to go to the next room? Enter 'yes' or 'no': {ENDC}").lower().strip() == 'yes':    
    
        # ALVARO: introducing a linebreak() for user readibility while playing.
                     
        linebreak()
        play_room(next_room)
        
    else:
       # ALVARO: introducing a linebreak() for user readibility while playing.
        linebreak()
        play_room(current_room)    
        
        
           
        
        
```
