```python
# Statistical Tests:

# Use chi-square to test independency between demographic variable and screen time reduction strategies. 

"""Null Hypothesis = There is no association between age, gender, occupation, 
                    and the choice of strategies to reduce screen time.
                    
This null hypothesis implies that the distribution of strategies is independent of the demographic variables 
(age, gender, occupation). In other words, the choice of strategies does not depend on or vary 
systematically with different demographic characteristics."""


alpha = 0.05

contingency_table = pd.crosstab([df_screentime['Age'], df_screentime['Gender'], df_screentime['Occupation']], 
                                df_screentime['Strategies'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}, p-value: {p_value}')


"""P-value is higher than 5%. It sugguests that we do not have enough evidence to reject 
the null hypothesis (fail to reject).

However in practical terms, this result suggests that there is not enough evidence to conclude that 
demographic characteristics significantly influence the choice of strategies used to reduce screen time"""
```

```python
# Now I want to check each demographic characteristics separately.

contingency_table = pd.crosstab(df_screentime['Age'], df_screentime['Strategies'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}, p-value: {p_value}')

```

```python
# Investigate association between awareness level and desire to reduce screen time 

# Null Hypothesis =  There is no association between awareness level and the desire to reduce 
# screen time among the surveyed individuals.

# Alternative Hypothesis = There is a significant association between awareness level and the desire
# to reduce screen time among the surveyed individuals.



from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df_screentime['Awareness'], df_screentime['Desire_to_Reduce_Screen_Time'])
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f"Chi-Square statistic: {chi2}")
print(f"P-value: {p}")

if p < 0.05:
    print("There is a significant association between awareness level and the desire to reduce screen time.")
else:
    print("There is no significant association between awareness level and the desire to reduce screen time.")
```

```python

countries = ['South Africa', 'Philippines', 'Brazil', 'Colombia', 'Argentina', 'Malaysia', 'Thailand', 'Mexico', 'Indonesia',
             'UAE', 'Taiwan', 'Saudi Arabia', 'Egypt', 'Turkey', 'Portugal', 'Russia', 'Israel', 'Singapore', 'India', 'Global Average']

proportion_streaming_tv = [96.6, 98.3, 97.7, 95.5, 96.7, 95.5, 97.1, 97.9, 97.1, 93.9, 94.8, 95.9, 91.2, 96.4, 84.7, 84.5, 95.7, 91.5, 97.2, 93.5]
proportion_playing_video_games = [90.2, 96.4, 89.3, 86.9, 86.6, 90.1, 94.7, 91.2, 94.5, 90.3, 91.6, 91.4, 86.3, 91.5, 81.2, 78.6, 75.2, 84.5, 92, 83.6]

daily_time_video_games = [63, 94, 70, 66, 61, 73, 118, 91, 79, 89, 53, 113, 123, 66, 38, 30, 19, 71, 71, 72]
daily_time_social_media = [223, 246, 181, 206, 61, 182, 178, 191, 79, 194, 124, 173, 123, 66, 28, 27, 58, 71, 81, 132]

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))

# Bar width
bar_width = 0.4
index = np.arange(len(countries))

# Bar for Proportion Streaming TV
bar_streaming_tv = ax.barh(index - bar_width/2, proportion_streaming_tv, bar_width, label='Streaming TV', color='skyblue')

# Bar for Proportion Playing Video Games
bar_playing_video_games = ax.barh(index + bar_width/2, proportion_playing_video_games, bar_width, label='Playing Video Games', color='lightcoral')

# Y-axis label
ax.set_xlabel('Proportion (%)', fontsize=12)

# Y-axis ticks
ax.set_yticks(index)
ax.set_yticklabels(countries, fontsize=10)

# Add percentages on top of bars
for i, (streaming_tv, playing_video_games) in enumerate(zip(proportion_streaming_tv, proportion_playing_video_games)):
    ax.text(streaming_tv + 1, index[i] - bar_width/2, f'{streaming_tv}%', ha='left', va='center', color='black', fontsize=8)
    ax.text(playing_video_games + 1, index[i] + bar_width/2, f'{playing_video_games}%', ha='left', va='center', color='black', fontsize=8)

# Create a second x-axis for time data
ax2 = ax.twiny()

# Line plot for Daily Time Spent on Video Games
line_color = 'gray'
ax2.plot(daily_time_video_games, index, color=line_color, marker='o', label='Video Games Time')
ax2.set_xlabel('Daily Time (mins)', color=line_color, fontsize=12)

# Line plot for Daily Time Spent on Social Media
ax2.plot(daily_time_social_media, index, color='darkgray', marker='o', linestyle='dashed', label='Social Media Time')

# X-axis ticks for the second x-axis
ax2.set_xticks(np.arange(0, max(max(daily_time_video_games), max(daily_time_social_media)) + 61, 60))
ax2.tick_params(axis='x', labelcolor=line_color, labelsize=10)

# Title
ax.set_title('Proportions and Daily Time Spent by Country', fontsize=14)

# Legends outside the chart area
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(1, 0.85))

# Display the chart
plt.tight_layout()
plt.show()
```
