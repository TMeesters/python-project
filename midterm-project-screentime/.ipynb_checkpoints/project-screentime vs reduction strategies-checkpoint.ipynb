```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Your awareness mapping
awareness_mapping = {
    'Not aware at all': 1,
    'Slightly aware': 2,
    'Moderate aware': 3,
    'Very aware': 4,
    'Extremely aware': 5
}

# Replace values using the mapping
df_screentime['Awareness'] = df_screentime['Awareness'].replace(awareness_mapping)

# Replace NaN values with -1 as a placeholder
df_screentime['Awareness'].fillna(-1, inplace=True)

# Convert the 'Awareness' column to numeric
df_screentime['Awareness'] = pd.to_numeric(df_screentime['Awareness'], errors='coerce')

# Convert 'Avg_hours' column to numeric
df_screentime['Avg_hours'] = pd.to_numeric(df_screentime['Avg_hours'], errors='coerce')

# Check if there are NaN values in 'Avg_hours'
print("NaN values in Avg_hours:", df_screentime['Avg_hours'].isna().sum())

# Define the order of awareness levels
awareness_order = ['Not aware at all', 'Slightly aware', 'Moderate aware', 'Very aware', 'Extremely aware']

# Create a bar plot
plt.figure(figsize=(10, 6))  
ax = sns.barplot(x='Awareness', y='Avg_hours', data=df_screentime, order=awareness_order, palette='pastel', ci=None)
plt.xlabel('Awareness Level')
plt.ylabel('Average Screen Time (Avg_hours)')
plt.title('Average Screen Time by Awareness Level')

# Annotate each bar with the count value
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', xy=(p.get_x() + p.get_width() / 2, height),
                xytext=(0, 5), textcoords='offset points',
                ha='center', va='baseline', fontsize=10, color='black')

plt.show()


```

```python
# Statistical Tests:

# Use chi-square to test independency between demographic variable and screen time reduction strategies. 

"""Null Hypothesis = There is no association between age, gender, occupation, 
                    and the choice of strategies to reduce screen time.
                    
This null hypothesis implies that the distribution of strategies is independent of the demographic variables 
(age, gender, occupation). In other words, the choice of strategies does not depend on or vary 
systematically with different demographic characteristics."""



alpha = 0.05


contingency_table = pd.crosstab([df_screentime['Age'], df_screentime['Gender'], df_screentime['Occupation']], df_screentime['Strategies'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}, p-value: {p_value}')


"""P-value is higher than 5%. It sugguests that we do not have enough evidence to reject 
the null hypothesis (fail to reject).

However in practical terms, this result suggests that there is not enough evidence to conclude that 
demographic characteristics significantly influence the choice of strategies used to reduce screen time"""
```

```python
# Now I want to check each demographic characteristics separately.

contingency_table = pd.crosstab(df_screentime['Age'], df_screentime['Strategies'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}, p-value: {p_value}')

```

```python
# There are 2 missing value in average screen time. I will use median to replace the null value. 
df_screentime['Avg_hours'].fillna(df_screentime['Avg_hours'].median(), inplace=True)
```

```python
# Statistical test to check association between high average screen time and and desire to reduce screen time.


from scipy.stats import mannwhitneyu


avg_hours_desire_yes = df_screentime[df_screentime['Desire_to_Reduce_Screen_Time'] == 'Yes']['Avg_hours']
avg_hours_desire_no = df_screentime[df_screentime['Desire_to_Reduce_Screen_Time'] == 'No']['Avg_hours']


statistic, p_value = mannwhitneyu(avg_hours_desire_yes, avg_hours_desire_no)


print(f"Mann-Whitney U statistic: {statistic}")
print(f"P-value: {p_value}")


if p_value < 0.05:
    print("There is a significant difference in average screen time between those who desire to reduce and those who don't.")
else:
    print("There is no significant difference in average screen time between the two groups.")
```

```python
"""Mann-Whitney U test compares the average screen time between two groups: 
1.individuals who desire to reduce screen time ('Yes') 
2. who do not ('No')

avg_hours_desire_yes: Average screen time for individuals who desire to reduce screen time.
avg_hours_desire_no: Average screen time for individuals who do not desire to reduce screen time.
"""
```
